Dropping the lat and log  557-546


```python
df = df.dropna(subset=["Lat", "Long"])
```

```python
df.shape
```

```python
df["Area"].fillna("Unknown", inplace=True)
```

```python
df["Address Type"].fillna("Others", inplace=True)
```

```python
df.isnull().sum()
```

```python
import numpy as np

df["Area"] = df["Area"].replace(["", " ", "NA", "null"], np.nan)

df["Area"].fillna("Unknown", inplace=True)
```

```python
df.isnull().sum()
```

```python
df[df["Area"].isnull()]
```

```python
df["Area"] = df.groupby("City")["Area"].transform(
    lambda x: x.fillna(x.mode()[0] if not x.mode().empty else "Unknown")
)
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
df.columns
```

```python
df.duplicated().sum()
```

Standardize city names

Fixes:

"bengaluru", "Bangalore", "BENGALURU"

(Optional manual mapping if needed)

```python
df["city"] = df["city"].str.strip().str.title()
```

```python
df["area"] = df["area"].str.lower().str.strip().str.replace(r"\s+", " ", regex=True)
```

```python
df.head(2)
```

```python
# //Extract Pincode

import re

def extract_pincode(area):
    match = re.search(r"\b\d{6}\b", str(area))
    return match.group(0) if match else None

df["pincode"] = df["area"].apply(extract_pincode)
```

```python
df["pincode"].value_counts().head()
```

```python
df["pincode"].value_counts()
```

```python
# Pincode extraction from address yielded limited results due to inconsistent address formats.

# Therefore, latitude and longitude were used as the primary features for spatial analysis, 
# as they provide more complete and reliable location information.



df.head()
```

```python
df["lat"] = pd.to_numeric(df["lat"], errors="coerce")
df["long"] = pd.to_numeric(df["long"], errors="coerce")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
# Remove Invalid Coordinates

# India roughly lies within:

# Latitude: 6 to 38
# Longitude: 68 to 98

df = df[(df["lat"] >= 6) & (df["lat"] <= 38)]
df = df[(df["long"] >= 68) & (df["long"] <= 98)]
```

```python
df.shape
```

```python
df.to_csv("../data/processed/cleaned_data.csv", index=False)
```
